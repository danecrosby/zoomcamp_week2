```python
import os
import pandas as pd

# Directory containing the zipped taxi files
directory = 'G:\zoomcamp\week3\mage-zoomcamp'

# Select which months to read and concat
start_month = 10
end_month = 12

# Initialize an empty list to store DataFrames
dfs = []

# Datatypes used for green taxi data
taxi_dtypes = {
                    'VendorID': pd.Int64Dtype(),
                    'passenger_count': pd.Int64Dtype(),
                    'trip_distance': float,
                    'RatecodeID':pd.Int64Dtype(),
                    'store_and_fwd_flag':str,
                    'PULocationID':pd.Int64Dtype(),
                    'DOLocationID':pd.Int64Dtype(),
                    'payment_type': pd.Int64Dtype(),
                    'fare_amount': float,
                    'extra':float,
                    'mta_tax':float,
                    'tip_amount':float,
                    'tolls_amount':float,
                    'improvement_surcharge':float,
                    'total_amount':float,
                    'congestion_surcharge':float
                }

parse_dates = ['lpep_pickup_datetime', 'lpep_dropoff_datetime']

# Loop through each file using the range function
for i in range(start_month, end_month + 1):
    filename = f'green_tripdata_2020-{i}.csv.gz'
    file_path = os.path.join(directory, filename)
    print(f'Reading and adding {file_path} to the dataframe')
    # Read the CSV file into a DataFrame and append to the list
    dfs.append(pd.read_csv(file_path, compression="gzip", dtype=taxi_dtypes, parse_dates=parse_dates))

# Concatenate all DataFrames in the list into a single DataFrame
concatenated_df = pd.concat(dfs, ignore_index=True)

# Display the concatenated DataFrame
concatenated_df.head(10)
```
