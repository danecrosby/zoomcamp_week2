```python
import pandas as pd
import gzip

# Path to the gzipped CSV file
file_path = 'green_tripdata_2020-10.csv.gz'

# Open the gzipped file
with gzip.open(file_path, 'rt') as file:
    # Load the CSV file into a Pandas DataFrame
    df = pd.read_csv(file)

# Display the DataFrame
df.head(10)
```
